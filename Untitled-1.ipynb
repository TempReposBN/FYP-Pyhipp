```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x) / 10  # Scaled to avoid large values

# Create a figure and a single subplot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot multiple lines on the same subplot
ax.plot(x, y1, label='Sine', color='r', linestyle='-')  # Line 1
ax.plot(x, y2, label='Cosine', color='g', linestyle='--')  # Line 2
ax.plot(x, y3, label='Tangent', color='b', linestyle='-.')  # Line 3

# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Multiple Lines in One Subplot')

# Add a legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Number of subplots
num_plots = 6

# Calculate grid size
rows = int(np.ceil(np.sqrt(num_plots)))
cols = int(np.ceil(num_plots / rows))

# Create figure and axes
fig, axs = plt.subplots(rows, cols, figsize=(12, 8))

# Flatten the array of Axes objects for easy iteration
axs = axs.flatten()

# Plot data in each subplot
for i in range(num_plots):
    axs[i].plot(np.linspace(0, 10, 100), np.sin(np.linspace(0, 10, 100) + i))
    axs[i].plot(np.linspace(0, 10, 100), np.sin(np.linspace(0, 10, 100) + i*2))
    axs[i].set_title(f'Plot {i + 1}')

# Remove any empty subplots if the grid is larger than the number of plots
for j in range(num_plots, len(axs)):
    fig.delaxes(axs[j])

# Adjust the layout
fig.tight_layout(pad=3.0)

# Show the figure
plt.show()
```

```python
import pandas as pd
from statsmodels.multivariate.manova import MANOVA

# Example data
data = {
    'Treatment': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Blood_Pressure': [120, 125, 130, 135, 140, 145, 150, 155, 160],
    'Cholesterol': [200, 210, 220, 230, 240, 250, 260, 270, 280]
}

df = pd.DataFrame(data)

# Perform MANOVA
manova = MANOVA.from_formula('Blood_Pressure + Cholesterol ~ Treatment', data=df)
print(manova.mv_test())
```
